```python
%matplotlib inline
from pylab import *
from glob import glob
from image_defs import *
import os
from Game import Board
```

```python
board_picture_folder="Images/2024-12-5 - training board images"
square_images_folder="Images/2024-12-5 - training squares"
image_to_use_for_corners="Images/2024-12-5 - training board images/demopic2.jpg"

corners= array([[89., 39.], 
                 [1105., 23.], 
                 [1198., 590.], 
                 [6., 593.]], dtype=float32) 
```

```python
filename=image_to_use_for_corners
```

```python
image=imread(filename)
subplot(2,1,1)
imshow(image)

image=image[300:900,:1200]
im3=straighten_image(image,corners)


subplot(2,1,2)
imshow(im3)
```

```python
state=Board('12122/11212/12212/21121')
squares=get_board_squares_from_image(im3,state.shape,middle_pixels=20)  # <=== check this
```

```python
nr,nc=state.shape
plot_count=0
figure(figsize=(10,10))
for r in range(nr):
    for c in range(nc):

        subplot(nr,nc,plot_count+1)
        imshow(squares[plot_count])

        shape=squares[plot_count].shape
        piece=state.board[plot_count]
        
        title(f"Piece {piece}")
        if c==0:
            ylabel(shape[0])

        if r==nr-1:
            xlabel(shape[1])
        gca().set_xticklabels([])
        gca().set_yticklabels([])

        
        plot_count+=1  
```
